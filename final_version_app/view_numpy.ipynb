```python
import os
import numpy as np
```

```python
def decode_numpy_array(arr):
    """Пытается расшифровать содержимое numpy массива."""
    print("\nФорма массива:", arr.shape)
    print("Тип данных:", arr.dtype)
    
    # Примеры расшифровки в зависимости от типа данных
    if np.issubdtype(arr.dtype, np.number):
        print("Это числовой массив.")
        print("Минимум:", np.min(arr))
        print("Максимум:", np.max(arr))
        print("Среднее:", np.mean(arr))
    elif np.issubdtype(arr.dtype, np.str_):
        print("Это строковый массив.")
        unique_values = np.unique(arr)
        print("Уникальные значения:", unique_values)
    elif np.issubdtype(arr.dtype, np.bool_):
        print("Это булев массив.")
        print("Количество True:", np.sum(arr))
    else:
        print("Тип данных не распознан для автоматической расшифровки.")

def main():
    folder_path = "./generated_db/numpy"
    
    if not os.path.exists(folder_path):
        print(f"Папка {folder_path} не существует!")
        return
    
    print(f"Чтение numpy массивов из {folder_path}...\n")
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".npy"):
            file_path = os.path.join(folder_path, filename)
            print(f"\n=== Файл: {filename} ===")
            
            try:
                # Загрузка массива
                array = np.load(file_path, allow_pickle=True)
                
                # Вывод содержимого
                print("Содержимое:")
                print(array)
                
                # Попытка расшифровки
                decode_numpy_array(array)
                
            except Exception as e:
                print(f"Ошибка при чтении файла {filename}: {e}")

if __name__ == "__main__":
    main()
```
